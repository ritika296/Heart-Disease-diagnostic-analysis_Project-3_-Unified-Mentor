```python
sns.countplot(x="target",data=data)  #countplot to check how many people have heart disease
plt.xticks(([0,1]),['Less chance of heart attack', 'More chance of heart attack'])
plt.xlabel('Target',fontsize=30)
plt.ylabel('Count',fontsize=30)
plt.show()
```

**Gender Count based on heart disease**

```python
sns.countplot(x='sex',hue='target',data=data)
plt.xticks(([1,0]),['Male','Female'])
plt.legend(labels=['No-Disease', 'Disease'])
plt.xlabel('Gender',fontsize=20)
plt.ylabel('Count',fontsize=20)
plt.title('Gender Count based on heart disease',fontsize=15)
plt.xticks([1,0],['Male','Female'])
plt.show
```

**find count of male and female in this dataset**

```python
data['sex'].value_counts()
```

```python
data["sex"].value_counts()     #Find count of Male and Female in this Dataset   
sns.countplot(x = "sex", data = data) 
plt.xticks(([0,1]),['Female', 'Male'])
plt.xlabel('',fontsize=25)
plt.ylabel('Count',fontsize=25)
plt.title('Count of Male & Female',fontsize=25)
plt.show()
```

**Count plot to check how many people have heart disease**

```python
sns.countplot(x="target",data=data)  
plt.xticks(([0,1]),['Less chance of heart attack', 'More chance of heart attack'])
plt.xlabel('Target',fontsize=25)
plt.ylabel('Count',fontsize=25)
plt.show()
```

**Chest Pain by Targeted People**

```python
sns.countplot(data=data,x='cp',hue='target',palette='CMRmap')
plt.legend(labels=['No-Disease', 'Disease'])
plt.xticks(([0,1,2,3]),['typical angina','atypical angina', 'non-anginal pain','asymptomatic'])
plt.xlabel('Chest Pain',fontsize=10)
plt.ylabel('Count',fontsize=10)
plt.title('Chest Pain by Targeted People',fontsize=10)
plt.show()
```

**Chest Pain by Age**

```python
sns.lineplot(x='cp',data=data,y='age',hue='sex')
plt.legend(labels=['Male','Female'])
sns.set(rc={'figure.figsize':(15,5)})
plt.xlabel('Chest Pain',fontsize=20)
plt.ylabel('Age',fontsize=20)
plt.title('Chest Pain by Age',fontsize=20)
plt.show()
```

**Fasting blood sugar(FBS) by Targeted People**

```python
sns.countplot(data=data,x='fbs',hue='target')
plt.legend(labels=['No-Disease', 'Disease'])# fasting blood sugar distribution according
plt.xlabel('Blood Sugar',fontsize=30)
plt.ylabel('Count',fontsize=30)
plt.title('FBS by Targeted People',fontsize=30)
plt.show
```

**Trestbps**

```python
sns.histplot(x='trestbps',data=data,color='black')
plt.xlabel('Trestbps',fontsize=10)
plt.ylabel('Count',fontsize=10)
plt.title('Trestbps',fontsize=10)
plt.show
```

**Age count by heart disease**

```python
sns.countplot(x='age',data=data,hue='target')
plt.xlabel('Age',fontsize=10)
plt.ylabel('Count',fontsize=10)
plt.title('Age count by heart disease',fontsize=10)
plt.show
```

**Cholesterol**

```python
sns.histplot(x='chol',data=data,color='r') # show distribution of serum cholesterol
plt.xlabel('Cholesterol',fontsize=10)
plt.ylabel('Count',fontsize=10)
plt.title('Cholesterol',fontsize=10)
plt.show
```

```python
g = sns.FacetGrid(data, hue="sex", aspect=4)
g.map(sns.kdeplot, 'trestbps', shade=True)
g.add_legend()
plt.show()
```
